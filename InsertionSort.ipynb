- Start with an empty left hand of playing cards and the cards face down on the table
- I Then remove one card at a time from the table, and insert it into the correct position in the left hand
- I To find the correct position for a card, compare it with each of the cards already in the hand, from right to left

```python
import random
```

### Intento sin ver soluci칩n

```python
def Sort(array):
    for j in range(len(array)-1):   # se utiliza la longitud de (array-1) 
                                    # para que cuando llegue al ultimo elemento de la lista no salga que no existe 
                
                valorSiguiente = array[j+1] # se conserva el siguiente valor de la lista
                
                if valorSiguiente < array[j]: 
                    array[j+1] = array[j] 
                    array[j] = valorSiguiente   # si el valor siguiente es menor, el menor se mueve 
                                                # a la posicion anterior
    return array

def InsertionSort(array):

    for i in range(len(array)-1): 
        for j in range(len(array)-1):
            if array[j] < array[i]: # si encuentra que hay algun valor menor que otro y 
                                    # no est치 en la posici칩n correcta, vuelve a repetir el algoritmo 
                                    # para colocarlo correctamente.
                Sort(array)
              
    return array
```

```python
A = random.sample(range(1, 50), 10) # Array
print ("Random number list is : " +  str(A))

print("Sorted list is : " + str(InsertionSort(A)))
```

### Intento con soluci칩n en mano

```python
def InsertionSort(array):
    for j in range(len(array)):
        key = array[j] # se guarda el valor siguiente
        i = j - 1
        while i >= 0 and array[i] > key: # mientras array[i] sea mayor que la llave
            array[i + 1] = array[i] # se intercambian lugares ambos valores
            i = i - 1 
        array[i + 1] = key # el siguiente valor se hace la siguiente llave
    return array
```

```python
A = random.sample(range(1, 50), 10) # Array
print ("Random number list is : " +  str(A))

print("Sorted list is : " + str(InsertionSort(A)))
```
